```python
import pandas as pd
import matplotlib.pyplot as plt

# read both data
df1 = pd.read_csv('OutcomeProblem_3.txt')
df2 = pd.read_csv('OutcomeArea_3.txt')

print(df1)
print(df2)
```

```python
#create figure
fig, ax = plt.subplots(figsize=(23, 10))

#scatter plot both data
ax.scatter(df1["Problem"], df1["Outcome"],color='red', alpha = 0.3, s = df1["Value"]*100,  label="Class Problem")
ax.scatter(df2["Area"], df2["Outcome"],color='blue', alpha = 0.3, s = df2["Value"]*100, label ="Research Area")
#ax.set_ylabel("Outcome", size=14)

#put value as dots anotation
for i, txt in enumerate(df1["Value"]):
    ax.annotate(txt, (df1["Problem"][i], df1["Outcome"][i]), ha='center', va='center', fontsize=13)  
for i, txt in enumerate(df2["Value"]):
    ax.annotate(txt, (df2["Area"][i], df2["Outcome"][i]), ha='center', va='center', fontsize=13)
    
#add grid
ax.grid(alpha =0.2, linestyle='--', linewidth=2)


#adjust thick in x axis
ax.axis(ymin = -0.8, ymax = 5.5, xmin = -0.5, xmax = 5.7)

#put y axis to the center and remove right and top axis line
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

#set x axis ticks
ax.set_xticks(pd.concat([df1["Problem"],df2["Area"]]).unique())

#set the position and plot the legends
#plt.legend(bbox_to_anchor=(1,0.1),loc="lower right",markerscale=0.3)

#change font size
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontname('Arial')
    label.set_fontsize(14)
    label.set_fontweight('bold')
    



#adjust the space between thick
ax.margins(x=0.8, y=0.2)

#show!
plt.show()
```
